```python
# Final program code 
import calendar

class Caregiver:
    def __init__(self, name, phone, email, pay_rate=20, is_paid=True):
        self.name = name
        self.phone = phone
        self.email = email
        self.pay_rate = pay_rate
        self.is_paid = is_paid  #family members arent paid
        self.hours_worked = 0
        self.availability = {'AM': 'unavailable', 'PM': 'unavailable'}
    
    #sets each users availibility
    def set_availability(self, shift, availability):
        if shift in self.availability:
            self.availability[shift] = availability
    
    #records hours
    def record_hours(self, hours):
        self.hours_worked += hours
    
    #record pay
    def weekly_pay(self):
        return self.hours_worked * self.pay_rate

    def reset_hours(self):
        self.hours_worked = 0

    def __repr__(self):
        return f"Caregiver({self.name}, {self.phone}, {self.email})"


class Schedule:
    def __init__(self, caregivers):
        self.caregivers = caregivers
        self.schedule = {}
    
    #add caregiver
    def add_caregiver(self, caregiver):
        self.caregivers.append(caregiver)
    
    #assign caregiver to shift
    def assign_shift(self, day, shift, caregiver):
        if caregiver.availability[shift] in ['preferred', 'available']:
            if day not in self.schedule:
                self.schedule[day] = {'AM': None, 'PM': None}
            if self.schedule[day][shift] is None:
                self.schedule[day][shift] = caregiver
                caregiver.record_hours(6)
            else:
                print(f"Shift {shift} on {day} is already filled.")
        else:
            print(f"{caregiver.name} is unavailable for {shift} on {day}")

    #generate schedule
    def generate_schedule(self, year, month):
        shifts = ["AM", "PM"]
        num_days = calendar.monthrange(year, month)[1]  #get the number of days in the month
        
        #initialize
        schedule = {}

        #assign caregivers based on their availability
        for day in range(1, num_days + 1):
            schedule[day] = {}

            #ensure all caregivers get a shift via a round robin approach
            #caregivers who already worked are moved to the bottom of the list
            for shift in shifts:
                available_caregivers = [caregiver for caregiver in self.caregivers if caregiver.availability[shift] in ['preferred', 'available']]

                if available_caregivers:
                    caregiver = available_caregivers[0]
                    schedule[day][shift] = caregiver
                    caregiver.record_hours(6)
                    self.caregivers.append(self.caregivers.pop(0))
                else:
                    schedule[day][shift] = "No caregiver available"

        return schedule
    
    #schedule HTML
    #used the example given mixed with some ChatGPT and Google to edit
    def display_schedule_as_html(self, schedule, year, month):
        """Display the generated schedule in an HTML table."""
        # Create the HTML structure
        html_schedule = f"""
        <html>
        <head>
            <title>Work Schedule for {calendar.month_name[month]} {year}</title>
            <style>
                table {{
                    border-collapse: collapse;
                    width: 100%;
                    margin: 20px 0;
                }}
                th, td {{
                    border: 1px solid black;
                    padding: 10px;
                    text-align: center;
                }}
                th {{
                    background-color: #f2f2f2;
                }}
                td {{
                    height: 100px;
                    vertical-align: top;
                }}
            </style>
        </head>
        <body>
            <h1>Work Schedule for {calendar.month_name[month]} {year}</h1>
            <table>
                <tr>
                    <th>Mon</th>
                    <th>Tue</th>
                    <th>Wed</th>
                    <th>Thu</th>
                    <th>Fri</th>
                    <th>Sat</th>
                    <th>Sun</th>
                </tr>
        """
        
        # Get the first weekday of the month and the total days
        first_weekday, num_days = calendar.monthrange(year, month)

        # Fill in the days of the month
        current_day = 1
        for week in range((num_days + first_weekday) // 7 + 1):
            html_schedule += "<tr>"
            for day in range(7):
                if (week == 0 and day < first_weekday) or current_day > num_days:
                    html_schedule += "<td></td>"  # Empty cell for days outside the month
                else:
                    # Add the day and the assigned shifts
                    shifts_for_day = schedule.get(current_day, {})
                    morning_shift = shifts_for_day.get("AM", "N/A")
                    afternoon_shift = shifts_for_day.get("PM", "N/A")

                    # Display only the name of the caregiver (not email/phone)
                    morning_shift_name = morning_shift.name if isinstance(morning_shift, Caregiver) else morning_shift
                    afternoon_shift_name = afternoon_shift.name if isinstance(afternoon_shift, Caregiver) else afternoon_shift

                    # Update the HTML to show only caregiver names
                    html_schedule += f"<td>{current_day}<br><b>AM:</b> {morning_shift_name}<br><b>PM:</b> {afternoon_shift_name}</td>"
                    current_day += 1
            html_schedule += "</tr>"

        # Close the table and HTML
        html_schedule += """
            </table>
        </body>
        </html>
        """
        
        # Write the HTML content to a file
        with open(f"work_schedule_{year}_{month}.html", "w") as file:
            file.write(html_schedule)

        print(f"HTML work schedule for {calendar.month_name[month]} {year} generated successfully!")

    def generate_pay_report(self):
        """Generate an HTML pay report that includes weekly and monthly totals."""
        # Initialize the HTML structure for the report
        report_html = f"""
        <html>
        <head>
            <title>Pay Report</title>
            <style>
                table {{
                    border-collapse: collapse;
                    width: 100%;
                    margin: 20px 0;
                }}
                th, td {{
                    border: 1px solid black;
                    padding: 10px;
                    text-align: center;
                }}
                th {{
                    background-color: #f2f2f2;
                }}
                td {{
                    height: 30px;
                }}
            </style>
        </head>
        <body>
            <h1>Pay Report</h1>
            <h2>Weekly Pay Breakdown</h2>
            <table>
                <tr>
                    <th>Caregiver</th>
                    <th>Weekly Pay (USD)</th>
                </tr>
        """

        # Calculate weekly pay for each caregiver and build the report
        total_weekly = 0
        for caregiver in self.caregivers:
            weekly = caregiver.weekly_pay()
            total_weekly += weekly
            report_html += f"<tr><td>{caregiver.name}</td><td>{weekly}</td></tr>"

        # Add total weekly pay
        report_html += f"""
                <tr>
                    <th>Total Weekly Pay</th>
                    <th>{total_weekly}</th>
                </tr>
            </table>

            <h2>Monthly Pay Estimate</h2>
            <p>Based on 4 weeks: <strong>{total_weekly * 4}</strong> USD</p>
        </body>
        </html>
        """

        # Save the HTML pay report to a file
        with open("pay_report.html", "w") as file:
            file.write(report_html)

        print("HTML pay report generated and saved as 'pay_report.html'.")
        
def main():
    # Create a list of caregivers
    caregiver1 = Caregiver("Alice Smith", "123-456-7890", "alice.smith@example.com", 20)
    caregiver2 = Caregiver("Bob Johnson", "321-654-0987", "bob.johnson@example.com", 20)
    caregiver3 = Caregiver("Charlie Brown", "234-567-8901", "cathy.brown@example.com", 20)
    caregiver4 = Caregiver("David Jones", "432-765-1098", "david.jones@example.com", 20)
    caregiver5 = Caregiver("Emma Davis", "345-678-9012", "emma.davis@example.com", 20)
    caregiver6 = Caregiver("Frank Wilson", "543-876-2109", "frank.wilson@example.com", 20)
    caregiver7 = Caregiver("Grace Lee", "456-789-0123", "grace.lee@example.com", 20)
    caregiver8 = Caregiver("Hank Miller", "654-987-3210", "hank.miller@example.com", 20)

    caregivers = [caregiver1, caregiver2, caregiver3, caregiver4, caregiver5, caregiver6, caregiver7, caregiver8]

    # Set some availability (you can set this as you want)
    caregiver1.set_availability("AM", "preferred")
    caregiver1.set_availability("PM", "available")
    caregiver2.set_availability("AM", "available")
    caregiver2.set_availability("PM", "unavailable")
    caregiver3.set_availability("AM", "unavailable")
    caregiver3.set_availability("PM", "available")
    caregiver4.set_availability("AM", "preferred")
    caregiver4.set_availability("PM", "preferred")
    caregiver5.set_availability("AM", "available")
    caregiver5.set_availability("PM", "available")

    # Create the schedule and assign caregivers
    schedule = Schedule(caregivers)

    # Ask for the year and month
    year = int(input("Enter the year (e.g. 2024): "))
    month = int(input("Enter the month (1-12): "))

    # Generate the work schedule
    schedule_for_month = schedule.generate_schedule(year, month)

    # Generate and display the work schedule as an HTML table
    schedule.display_schedule_as_html(schedule_for_month, year, month)

    schedule.generate_pay_report()
    
if __name__ == "__main__":
    main()
```
